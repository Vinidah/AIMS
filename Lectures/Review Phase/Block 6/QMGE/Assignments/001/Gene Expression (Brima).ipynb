```python
import numpy as np
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
```

```python
#### Question 2; Gene is introduced into a newly replicated cell
def model(M,t):
    beta = 0.2
    M = 2       # mRNA is constant, you can choose any value but note that this does not show the steady state.
    dmdt = beta*M
    return dmdt

t = np.linspace(0,10,100) 
M_t = odeint(model,1,t)
plt.plot(t,M_t,'r')
plt.xlabel('time ')
plt.ylabel('M (mRNA) ')
plt.title("Temporal concentration of mRNA")
plt.legend(["Level of mRNA"],loc='lower right')
plt.show()
```

```python
###question 4 
def model(P,t):
    beta = 2
    M = 0         # mRNA (M) is zero since there is  no coding for translation to protein
    dPdt = beta*M  #in the newly replicated cell lacking P
    return dPdt

t = np.linspace(0,10,100)
P_t = odeint(model,5,t)
plt.plot(t,P_t,'r-')
plt.title("Temporal concentration of protein")
plt.legend(["Concentration of protein"],loc='lower right')
plt.xlabel("time")
plt.ylabel("P")
plt.show()
```

```python
###question 4 
def model(P,t):
    beta = 2
    gamma = 3
    M = 10
    dPdt =  -gamma*P  #in the newly replicated cell lacking M
    return dPdt

t = np.linspace(0,10,100)
P_t = odeint(model,5,t)
plt.plot(t,P_t,'r')
plt.title("Temporal concentration of protein")
plt.legend(["Concentration of protein"],loc='upper right')
plt.xlabel("time")
plt.ylabel("P")
plt.show()
```

```python
###question 5 
def model(x,t):
    M,P = x
    beta = 2
    gamma_m = 1
    alpha = 2
    gamma_p = 2
    dMdt = 10 - gamma_m*M   # Here (M_0) is chosen as 10 it can be changed.
    dPdt = alpha*M - gamma_p*P
    return np.array([dMdt,dPdt])

ini = [0,0]
t = np.linspace(0,10,100)
MP_t = odeint(model,ini,t)
MAt = MP_t[:,0]
P_t = MP_t[:,1]


#plt.figure(figsize=(12,8))
#plt.subplot(1,2,1)
# plt.plot(t,MAt,t,P_t,'r') #plot of mRNA(M) and protein (P) against time
# plt.xlabel("time")
# plt.ylabel("M(t) and P(t)")
# plt.legend(['mRNA(M)','Protein(P)'])
# plt.title('Temporal concentration of M and P')
# plt.show()
#plt.subplot(1,2,2)
plt.plot(MAt,P_t,'y') # plot of mRNA (M) against protein (P)
plt.title('Concentration of M against P')
plt.ylabel("mRNA")
plt.xlabel("Protein ")
plt.show()
```

```python
###### question 8 
def model(x,t):
    M,P = x
    beta = 2
    gamma_m = 1
    alpha = 2
    k = 0.1       #we have a very small amount of repressor R in the cell
    k_d = 0.01
    gamma_p = 2
    dMdt = 10 - gamma_m*M - k*M  
    dPdt = alpha*M - gamma_p*P - k_d*P
    return np.array([dMdt,dPdt])

ini = [0,0]
t = np.linspace(0,10,100)
MP_t = odeint(model,ini,t)
MAt = MP_t[:,0]
Pt = MP_t[:,1]


plt.figure(figsize=(10,7))
plt.plot(t,MAt,t,Pt,'r')
plt.xlabel("time")
plt.ylabel("M(t) and P(t) ")
plt.legend(['mRNA(M)','Protein(P)'])
plt.title('Temporal concentration of M and P')

plt.show()
```

```python
### question 9  This considers huge values for k and k_d
def model(x,t):
    M,P = x
    beta = 2
    gamma_m = 1
    alpha = 2
    k = 20     #we have a lot of repressor R in the cell 
    k_d = 10
    gamma_p = 2
    dMdt = 10 - gamma_m*M - k*M
    dPdt = alpha*M - gamma_p*P - k_d*P
    return np.array([dMdt,dPdt])

ini = [0,0]
t = np.linspace(0,10,100)
MP_t = odeint(model,ini,t)
M_t = MP_t[:,0]
P_t = MP_t[:,1]


plt.figure(figsize=(10,7))
plt.plot(t,M_t,t,P_t,'r')
plt.xlabel("Time")
plt.ylabel("mRNA and Protein")
plt.legend(['mRNA(M)','Protein(P)'])
plt.title('Temporal concentration of M and P')
plt.show()
```
