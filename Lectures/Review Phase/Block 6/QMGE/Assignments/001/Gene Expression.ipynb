```python
import numpy as np
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
```

```python
#### Question 2; Gene is introduced into a newly replicated cell
def model(M,t):
    beta = 0.2
    M = 2          # mRNA is constant, you can choose any value but note that this does not show the steady state.
    dmdt = beta*M
    return dmdt

t = np.linspace(0,10,100) 
M_t = odeint(model,1,t)
plt.plot(t,M_t)
plt.title 
plt.show()
```

```python
###question 4 
def model(P,t):
    beta = 2
    M = 0         # mRNA (M) is zero since there is  no coding for translation to protein
    dPdt = beta*M  #in the newly replicated cell lacking P
    return dPdt

t = np.linspace(0,10,100)
P_t = odeint(model,5,t)
plt.plot(t,P_t)
plt.show()
```

```python
###question 4 
def model(P,t):
    beta = 2
    gamma = 3
    M = 10
    dPdt =  -gamma*P  #in the newly replicated cell lacking M
    return dPdt

t = np.linspace(0,10,100)
P_t = odeint(model,5,t)
plt.plot(t,P_t)
plt.show()
```

```python
###question 5 
def model(x,t):
    M,P = x
    beta = 2
    gamma_m = 1
    alpha = 2
    gamma_p = 2
    dMdt = 10 - gamma_m*M   # Here (M_0) is chosen as 10 it can be changed.
    dPdt = alpha*M - gamma_p*P
    return np.array([dMdt,dPdt])

ini = [0,0]
t = np.linspace(0,10,100)
MP_t = odeint(model,ini,t)
M_t = MP_t[:,0]
P_t = MP_t[:,1]


plt.figure(figsize=(7,10))
plt.subplot(2,1,1)
plt.plot(t,M_t,t,P_t) #plot of mRNA(M) and protein (P) against time
plt.xlabel("Time")
plt.ylabel("mRNA and Protein levels")
plt.legend(['mRNA(M) level','Protein(P) level'])
plt.title('Levels of M and P against time')

plt.subplot(2,1,2)
plt.plot(M_t,P_t) # plot of mRNA (M) against protein (P)
plt.title('Levels of M against P')
plt.ylabel("mRNA levels")
plt.xlabel("Protein ")
plt.show()
```

```python
###### question 8 
def model(x,t):
    M,P = x
    beta = 2
    gamma_m = 1
    alpha = 2
    k = 0.1       #we have a very small amount of repressor R in the cell
    k_d = 0.01
    gamma_p = 2
    dMdt = 10 - gamma_m*M - k*M  
    dPdt = alpha*M - gamma_p*P - k_d*P
    return np.array([dMdt,dPdt])

ini = [0,0]
t = np.linspace(0,10,100)
MP_t = odeint(model,ini,t)
M_t = MP_t[:,0]
P_t = MP_t[:,1]


plt.figure(figsize=(7,8))
plt.plot(t,M_t,t,P_t)
plt.xlabel("Time")
plt.ylabel("mRNA and Protein levels")
plt.legend(['mRNA(M) level','Protein(P) level'])
plt.title('Levels of M and P against time')

plt.show()
```
