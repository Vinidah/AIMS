## Exercise 1: Cluster Analysis (CA)

```r
dakar <- read.csv(file = './Dataset/Dakar.csv')
head(dakar)
```

```r
#We are filtering out August which has the highest average precipitation
dakar.august = dakar[dakar$Month == 'AUG',]
```

```r
#We are inputting the missing values
dakar.august[is.na(dakar.august)] <- 0
```

```r
#We are inputting the missing values
dakar.august.extracted <- dakar.august[,4:13]

#dakar.august.extracted[is.na(dakar.august.extracted )] <- 0

dakar.august.extracted <- replace(dakar.august.extracted, is.na(dakar.august.extracted), 0)
```

```r
# scaling the array

dakar.august.scaled=scale(dakar.august.extracted)
```

```r
dakar.august.scaled<- replace(dakar.august.scaled, is.na(dakar.august.scaled), 0)
```

```r


####### performing the clustering############
dakar.august.scaled.clus=hclust(dist(t(dakar.august.scaled)),method='ward.D2')

#####plotting the tree#################
plot(dakar.august.scaled.clus,hang=-6,col="blue")

#cutree(yy.clus,k=4)
```

```r
####### performing the clustering############
dakar.august.scaled.clus=hclust(dist(t(dakar.august.scaled)),method='average')

#####plotting the tree#################
plot(dakar.august.scaled.clus,hang=-6,col="brown")

#cutree(yy.clus,k=4)
```

```r
####### performing the clustering############
dakar.august.scaled.clus=hclust(dist(t(dakar.august.scaled)),method='single')

#####plotting the tree#################
plot(dakar.august.scaled.clus,hang=-6,col="red")

#cutree(yy.clus,k=4)
```

```r
# Cut tree into 4 groups
sub_grp <- cutree(dakar.august.scaled.clus, k = 4)
```

## Exercise 2: Principal Component Analysis (PCA)

```r
# calculating the PCA
#, center = TRUE,scale = TRUE
dakar.pca <- prcomp(dakar.august.extracted)

summary(dakar.pca)
```

```r
#ggbiplot(dakar.pca)
#ggbiplot(dakar.pca, labels=rownames(dakar.august.extracted))
```

```r
#Scree plot
#screeplot(dakar.pca,type='l',lwd=3,main='Scree Plot')

fviz_eig(dakar.pca)
```

```r
fviz_pca_ind(dakar.pca,
             col.ind = "cos2", # Color by the quality of representation
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE     # Avoid text overlapping
             )
```

```r
fviz_pca_var(dakar.pca,
             col.var = "contrib", # Color by contributions to the PC
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE     # Avoid text overlapping
             )
```

#### Access to the PCA results

```r
library(factoextra)
# Eigenvalues
eig.val <- get_eigenvalue(dakar.pca)
eig.val
  
```

```r
# Results for Variables
res.var <- get_pca_var(dakar.pca)
res.var$coord          # Coordinates
res.var$contrib        # Contributions to the PCs
res.var$cos2           # Quality of representation 
```

```r
# Results for individuals
res.ind <- get_pca_ind(dakar.pca)
res.ind$coord          # Coordinates
res.ind$contrib        # Contributions to the PCs
res.ind$cos2           # Quality of representation 
```

```r
#loadings

# Matching the STATISTICA ouput 

mm=array(NA,c(10,10))

for (i in 1:10){
mm[,i]=dakar.pca$rotation[,i]*dakar.pca$sdev[i]
}

rownames(mm)=names(dakar.august.extracted)
colnames(mm)=colnames(dakar.pca$rotation)
```

```r
print('#######################################')
print('##  FIRTS FIVE  LOADINGS              #')
print('#######################################')

print(mm[,1:5])


# Explained variance

ev=round(dakar.pca$sdev^2/sum(dakar.pca$sdev^2)*100,2)



print('#########################################')
print('##  EXPLAINED VARIANCE (IN %)           #')
print('#########################################')

print(ev)
```
