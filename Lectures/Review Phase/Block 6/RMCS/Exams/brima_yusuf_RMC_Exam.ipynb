```r
x <-  dataset[,1]
y <- dataset[,2]
m1 = lm(y~x)
plot(city,xlab = '', ylab='mm/month', main='Annual Precipitation in Cape Town',col="brown1")
abline(m1)
```

```r
city_d <- diff(city, lag = 1)


x <-  dataset[,1]
y <- dataset[,2]
m1 = lm(y~x)
plot(city_d,xlab = '', ylab='mm/month', main='Annual Precipitation in Cape Town detrended',col="brown1")
abline(m1)

 
```

## TASK 4: Time-series Analysis Part II (Spectral Analysis and Wavelet Analysis)

```r
### Autocorrelation plot
forecast::Acf(city_d, lag.max = 12)
```

```r
###################---------------------------###################
#################                                 ###############
#                       Time Series Analysis                    #
#                         Fourier Analysis                      #
#################                                 ###############
###################---------------------------###################
par(mfcol = c(1,1))
p <- periodogram(city_d)
```

```r
## converting the frequency to periods
period_city <- 1/p1$freq
city <- ts(city_d, start = c(1960), end = c(2019), frequency = 1)
par(mfcol = c(1,1))
plot(period_city, p$spec, type = 'b', xlim = c(2,13), xlab = 'period', ylab = 'power',main='Annual Precipitation in Cape Town')
```

```r
###################---------------------------###################
#################                                 ###############
#                       Time Series Analysis                    #
#             Wavelet Analysis and Wavelet Coherence            #
#################                                 ###############
###################---------------------------###################

## Wavelet
timelong=seq(1960, 2020-1/12, 1)  # indexing
wavelet1=wt(cbind(timelong,city),dj=0.1,mother="morlet",max.scale=16) #wavelet

## plot of wavelet
par(mfcol = c(1,1))
par(oma=c(0, 0, 0, 1), mar=c(5, 4, 4, 5) + 0.1)
xx=plot(wavelet1,  plot.cb = TRUE,lwd.coi = 1,col.coi="white",alpha.coi=0.5,col.sig="black",
        lwd.sig = 2, ncol = 768, tol = 0.95, plot.phase=F,ylab="Period(Year)", xlab = "Time(Year)",
         main='Annual Precipitation in Cape Town')
```
