```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def RK4(func,m,t):
    h = t[1] - t[0]
    Z = np.zeros((len(t),len(m)))
    Z[0] = m
    for n in range(0,len(t)-1):
        k1 = h *func(Z[n],t[n])
        k2 =  h * func(Z[n] + (h * (k1/2)),t[n] + (h/2))
        k3 = func(Z[n] + (h * (k2/2)),t[n] + (h/2))
        k4 =  func(Z[n] + (h * k3),t[n]+ h)
        Z[n+1] =  Z[n] + (1/6)* h * ( k1 + 2*k2 + 2*k3 + k4)
    return Z
```

## Exercise 1 i

```python
def f1(c,t):
    x,y = c
    return np.array([5*x + y,3*x + y])

t =  np.linspace(-10,20,3000)
n=10
C  = np.round(np.abs(np.random.randn(30,2) * n))
fig = plt.figure()
for i in range(C.shape[0]):
    res = RK4(f1,C[i],t)
    x,y =  res.T
    plt.plot(x,y)
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.title('Dynamic')
plt.show()
```

```python
def f2(c,t):
    x,y = c
    dx =  2*x -  5*y
    dy =  x - 2*y
    return np.array([dx,dy])

fig = plt.figure()
for i in range(C.shape[0]):
    res = RK4(f2,C[i],t)
    x,y =  res.T
    plt.plot(x,y)
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.title('Dynamic')
plt.show()
```

```python
def f_n(c,t):
    x,y = c
    return np.array([y  -  x**2,3*x - 2*y])

fig = plt.figure()
for i in range(C.shape[0]):
    res = RK4(f_n,C[i],t)
    x,y =  res.T
    plt.plot(x,y)
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.title('Dynamic')
plt.show()
```

```python
def model(c,t):
    x,y = c
    dx  = -4 * t + t
    dy = t
    return np.array([dx,dy])

fig = plt.figure()
for i in range(C.shape[0]):
    res = RK4(model,C[i],t)
    x,y =  res.T
    plt.plot(x,y)
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.title('Dynamic')
plt.show()
```
