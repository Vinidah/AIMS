```python
%matplotlib inline
import numpy as np
import pylab as py
import matplotlib.pyplot as plt
```

```python
def RungeKutta4(func,M,t,H):
    arr = np.zeros((len(t),len(M)))
    arr[0] = M
    for n in range(0,len(t)-1):
        k1 = H *func(t[n],arr[n])
        k2 =  H * func(t[n] + (H/2),arr[n] + (H * (k1/2)))
        k3 = func(t[n] + (H/2),arr[n] + (H * (k2/2)))
        k4 =  func(t[n]+ H,arr[n] + (H * k3))
        arr[n+1] =  arr[n] + (1/6)* H * ( k1 + 2*k2 + 2*k3 + k4)
    return arr
```

## Exercise 1.1

```python
def func1(t,k):
    x,y = k
    return np.array([5*x + y,3*x + y])

time =  np.linspace(-1,1,250)
H =  time[1] - time[0]

for x0 in np.linspace(-1,1,20):
    for y0 in np.linspace(-1,1,20):
        M =  [x0,y0]
        arr = RungeKutta4(func1,M,time,H)
        x =  arr[:,0]
        y = arr[:,1]
        plt.plot(x,y)

plt.xlabel('x(t)', fontsize=4)
plt.ylabel('y(t)', fontsize=4)
plt.tick_params(labelsize=4)
plt.xlim(-1, 1)
plt.ylim(-1, 1);
xPos,yPos = np.mgrid[-1:1:10j, -1:1:10j]
u  =  5*xPos + yPos
v  =  3*xPos + yPos
py.quiver(xPos, yPos, u, v, color="green")
plt.show()
```

```python
def func2(t,k):
    x,y = k
    dx =  2*x -  5*y
    dy =  x - 2*y
    return np.array([dx,dy])


for x0 in np.linspace(-1,1,10):
    for y0 in np.linspace(-1,1,10):
        M =  [x0,y0]
        arr = RungeKutta4(func2,M,time,H)
        x =  arr[:,0]
        y = arr[:,1]
        plt.plot(x,y)

plt.xlabel('x(t)', fontsize=4)
plt.ylabel('y(t)', fontsize=4)
plt.tick_params(labelsize=4)
plt.xlim(-1, 1)
plt.ylim(-1, 1);
xPos,yPos = np.mgrid[-1:1:10j, -1:1:10j]
u  =  2*xPos  - 5*yPos
v  =  xPos + 2*yPos
py.quiver(xPos, yPos, u, v, color="green")
plt.show()
```
