```python
%matplotlib inline
import numpy as np
import pylab as p
np.seterr(all='ignore')
import matplotlib.pyplot as plt
```

```python
def RK4(func,m,t):
    H = t[1] - t[0]
    V = np.zeros((len(t),len(m)))
    V[0] = m
    for n in range(0,len(t)-1):
        k1 = H *func(V[n],t[n])
        k2 =  H * func(V[n] + (H * (k1/2)),t[n] + (H/2))
        k3 = func(V[n] + (H * (k2/2)),t[n] + (H/2))
        k4 =  func(V[n] + (H * k3),t[n]+ H)
        V[n+1] =  V[n] + (1/6)* H * ( k1 + 2*k2 + 2*k3 + k4)
    return V
```

## Exercise 1 i

```python
def f1(c,t):
    x,y = c
    return np.array([5*x + y,3*x + y])

t =  np.linspace(-10,20,3000)

fig = plt.figure(figsize=(10,10))
for x0 in np.linspace(-1,1,20):
    for y0 in np.linspace(-1,1,20):
        C =  [x0,y0]
        res = RK4(f1,C,t)
        x,y =  res.T
        plt.plot(x,y)

plt.xlabel('x(t)', fontsize=4)
plt.ylabel('y(t)', fontsize=4)
plt.tick_params(labelsize=4)
plt.xlim(-1, 1)
plt.ylim(-1, 1);
# Plot the vectorfield.
X,Y = np.mgrid[-1:1:10j, -1:1:10j]
u  =  5*X + Y
v  =  3*X + Y
p.quiver(X, Y, u, v, color="blue")
plt.title('Dynamical System')
plt.show()
```

```python
def f2(c,t):
    x,y = c
    dx =  2*x -  5*y
    dy =  x - 2*y
    return np.array([dx,dy])

fig = plt.figure(figsize=(10,10))
for x0 in np.linspace(-1,1,20):
    for y0 in np.linspace(-1,1,20):
        C =  [x0,y0]
        res = RK4(f2,C,t)
        x,y =  res.T
        plt.plot(x,y)

plt.xlabel('x(t)', fontsize=4)
plt.ylabel('y(t)', fontsize=4)
plt.tick_params(labelsize=4)
plt.xlim(-1, 1)
plt.ylim(-1, 1);
# Plot the vectorfield.
X,Y = np.mgrid[-1:1:10j, -1:1:10j]
u  =  2*X  - 5*Y
v  =  X + 2*Y
p.quiver(X, Y, u, v, color="blue")
plt.title('Dynamical System')
plt.show()
```
