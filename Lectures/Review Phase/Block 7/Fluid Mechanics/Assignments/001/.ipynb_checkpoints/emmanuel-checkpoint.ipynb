```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Question (4)  (a) (i) Under-damped


```python
def acc(x,beta,v):
    return -x - (beta * v)


def ver(c,d,dt,x0,v0,a,beta):
    t =  np.arange(c,d+dt,dt)
    N =  len(t)
    x  =  np.zeros(N)
    v =  np.zeros(N)
    x[0] = x0
    v[0] = v0
    for n in range(N -1):
        a_n = a(x[n], beta, v[n])
        x[n+1] = x[n] + (dt * v[n]) + 0.5* (dt**2 * a_n)
        v_hat =  v[n] + (0.5 * dt) *( a_n + a( x[n+1],beta,v[n] + (dt*a_n)))
        v[n+1] = v[n] + (0.5*dt) * (a_n + a(x[n+1], beta, v_hat))
    return t,x,v
```

```python
c = 0
d = 20
x0 = 1
v0 = 1
beta = 0.4
dt = 0.01
betas =  np.linspace(0.1,1.9)
f = plt.figure(figsize=(14,8))
for beta in betas:
    t,x,v = ver(c,d,dt,x0,v0,acc,beta)
    plt.plot(t,x)
plt.xlabel("time")
plt.ylabel("x(t)")
plt.show()
```

```python
fig = plt.figure(figsize=(10,8))
for beta in betas:
    t,x,v = ver(c,d,dt,x0,v0,acc,beta)
    plt.plot(x,v)
plt.xlabel(rf"x(t)")
plt.ylabel(rf"v(t)")
plt.show()
```

### Question (4)  (b)

$
\begin{align}
a_n  = -x - \beta \sin\left( v \right) 
\end{align}
$


```python
def acc(x,beta,v):
    a = -x + beta* np.sin(v)
    return a

c = 0
d = 20
x0 = 1
v0 = 0
beta = 0.6
dt = 0.01
betas =  np.linspace(0.1,1.9)
f = plt.figure(figsize=(14,8))
for beta in betas:
    t,x,v = ver(c,d,dt,x0,v0,acc,beta)
    plt.plot(t,x)
plt.xlabel("time")
plt.ylabel("x(t)")
plt.show()
```
