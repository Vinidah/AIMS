```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def interest(c,n,i):
    si =  c*(1 + n*i)
    ci =  c*(1 + i)**n
    return np.array([si,ci])

def compute_interest(c,i,t,interest):
    res =  np.zeros((len(t),2))
    print(res.shape)
    for k in range(0,len(t)):
        res[k] =  interest(c,t[k],i)
    return res
```

```python
n = 80
t = np.arange(10,n+1,1)
i =  0.08
c =  100.0
res = compute_interest(c,i,t,interest)

plt.figure(figsize=(10,8))
plt.plot(t,res[:,0])
plt.plot(t,res[:,1])
plt.title("Interest Calculation")
plt.xlabel("Time (years)")
plt.ylabel("Interest (any currency)")
plt.legend(["Simple Interest","Compound Interest"])
plt.show()
```
