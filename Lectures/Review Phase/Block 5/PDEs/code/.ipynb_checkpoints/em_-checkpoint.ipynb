```python
import numpy as np
from scipy.signal import square
import matplotlib.pyplot as plt
```

```python
def model(x,N):
    a = 0
    for n in range(1,N+1):
        a += np.sin(((2*n) -1 )*x)/((2*n) -1)
    result = (a*4)/np.pi
    return(result)
vev_func = np.vectorize(model)
```

#### Exercise 3b

```python
x=np.arange(0,np.pi,0.01)


N = 5


y_1 = vev_func(x,N)

N  = 10
y_2 = vev_func(x,N)

N = 20
y_3 = vev_func(x,N)
plt.plot(x/np.pi, y_1, 'g')
plt.plot(x/np.pi, y_2, 'r')
plt.plot(x/np.pi, y_3,'y')
plt.legend(['N=5',"N=10","N=20"], loc='lower left')
plt.ylabel('fN(x)')
plt.xlabel("x/π")
plt.show()
```

```python
x=np.arange(0,0.1,0.001)
N = 20


y_1 = vev_func(x,N)

N = 50
y_2 = vev_func(x,N)

N = 500
y_3 = vev_func(x,N)

plt.plot(x, y_1, 'g')
plt.plot(x, y_2, 'b')
plt.plot(x, y_3,'r')
plt.legend(['N = 20',"N = 50","N = 500"], loc='lower right')
plt.ylabel("f_N(x)")
plt.xlabel("x")
plt.show()
```
