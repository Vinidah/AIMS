```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    if( (x >= -np.pi  and x <0)  or ( x >= -3*np.pi  and x < -2*np.pi) or ( x>= np.pi and x < 2*np.pi)) :
        return -1
    else:
        return 1
```

```python
x =  np.linspace(-3*np.pi, 3*np.pi, 50)
```

```python
fv  = np.vectorize(f)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x,fv(x),'--go',linewidth=1.2)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Fourier Sine Seriers for $2\pi$ periodic function")
#plt.grid()
#plt.legend(["$2\pi$ periodic fourier"])
plt.legend(["$2\pi$ fourier"], loc='upper left')
plt.show()
```
