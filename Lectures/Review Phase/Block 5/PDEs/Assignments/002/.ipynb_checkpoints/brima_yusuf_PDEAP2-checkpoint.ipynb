```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square
from scipy.integrate import quad
```

```python
def f(x,N):
    a = 0
    for n in range(1,N+1):
        a += np.sin(((2*n) -1 )*x)/((2*n) -1)
    return (a*4)/np.pi

def ff(x):
    return 1
    
```

```python
x=np.arange(-np.pi,np.pi,0.001)
fv =  np.vectorize(ff)
y=square(x)


plt.figure(figsize=(10,6))
plt.plot(x/np.pi,y, 'b.')
plt.legend(['f(x)'])
plt.xlabel(r"$\frac{x}{x}$")
plt.grid()
plt.ylabel("y")
plt.show()
```
