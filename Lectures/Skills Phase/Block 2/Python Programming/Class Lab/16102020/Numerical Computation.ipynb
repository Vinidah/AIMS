```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x =  np.linspace(-2,8)

y =  (2*x**3) + (1/(x**2 +1))
```

```python
plt.plot(x,y)
plt.show()
```

```python
x =  np.linspace(-5.5,11.2)

y =  np.exp(-2*x) + np.log(4*x**2 +1) +  np.abs(x+2)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
