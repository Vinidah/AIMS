Let $\{ x_k\}$ be a partition of $[a,b]$ such that $a=x_0<x_1<\cdots<x_{N-1}<x_{N}=b$ and $H$ be the constant length of the $k$-th subinterval ($H = x_k - x_{k-1}$). Let us consider initial value problem

\begin{equation}\label{eul2}
  \begin{cases}
    y' = f(x,y),      & \quad \text{on } [a, b]\\\\
    y(a) = c,
  \end{cases}
\end{equation}
Let $\{ y_k\}$ be the approximate solution i.e.  $y_k\approx g(x_k)=g_k$ where $g$ is the exact solution.
1. Write a python function <b> EulerMethod </b> that takes $a,b,c,N,$ and $f$ and return array of all $x_k$ and $y_k$ of \eqref{eul2} using Euler method i.e.
$$ y_{k+1} = y_k + Hf(x_k,y_k) $$


```python
#definition for EulerMethod


def EulerMethod(a,b,c,N,f):
    x =  np.linspace(a,b,N+1)
    
    H =  x[1] - x[0]
    y =  np.zeros(N+1)
    y[a] = c
    for i in range(1,N+1):
        y[i] = y[i-1] + H*(f(x[i-1],y[i-1]))
    return x,y

```

2. Write a python function <b> RK2Method </b> that takes $a,b,c,N,$ and $f$ and return array of all $x_k$ and $y_k$ of (1) using second order Runge-Kutta  method i.e.
\begin{equation}
\begin{cases}
\alpha = f(x_k,y_k)\\\\
y_{k+1} = y_k + Hf(x_k+\dfrac{H}{2}, y_k + \dfrac{H}{2}\alpha)
\end{cases}
\end{equation}



```python

def  RK2Method(a,b,c,N,f):
    x =  np.linspace(a,b,N+1)
    
    H =  x[1] - x[0]
    y =  np.zeros(N+1)
    y[a] = c
    for i in range(1,N+1):
        y[i] = y[i-1] + H *(f((x[i-1] + H/2),(y[i-1]) + (H/2) * f(x[i-1],y[i-1])  ))
    return x,y

```

3. Let us consider the initial value problem 

\begin{equation}
  \begin{cases}
    y' = 2y + e^{2x},      & \quad \text{on } [0, 1]\\\\
    y(0) = 3,
  \end{cases}
\end{equation}
with exact solution $g(x) = (x+3)e^{2x}$. 

For $N=200$, use subplot to plot side by side
- the exact solution and the approximate solutions obtained with <b> EulerMethod </b> and <b> RK2Method </b>
- the absolute error between the exact solution and the approximate solutions obtained with <b> EulerMethod </b> and <b> RK2Method </b>

```python
f =  lambda x,y: 2*y + np.exp(2*x) #approximation function
g  =  lambda x: (x+3)*np.exp(2*x) #Actual function 
z = lambda f,g: abs(f-g) # error function 

a = 0
b = 1
c  = 3
N =  200

#EulerMethod Approximation Test 
x_Euler,y_Euler = EulerMethod(a,b,c,N,f) 
y_Euler_Exact = g(x_Euler)



#absolute error between the exact solution and the approximate solutions obtained with EulerMethod
error_Euler = z(y_Euler,y_Euler_Exact)


#Runge-Kutta Approximation Test 
x_RK,y_RK = RK2Method (a,b,c,N,f) 
y_RK_Exact = g(x_RK)

#absolute error between the exact solution and the approximate solutions obtained with RK2Method 
error_RK = z(y_RK,y_RK_Exact)

plt.subplot(2,1,1)
plt.plot(x_Euler,y_Euler)
plt.plot(x_Euler[::10],y_Euler_Exact[::10],'ro')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["Approx","Exact"])
plt.title("Function approximation with Euler Method")
plt.show()



plt.subplot(2,1,2)
plt.plot(x_RK,y_RK)
plt.plot(x_RK[::10],y_RK_Exact[::10],'ro')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["Approx","Exact"])
plt.title("Function approximation with Runge-Kutta")
plt.show()


plt.subplot(2,1,1)
plt.plot(x_Euler,error_Euler)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["Absolute Error"])
plt.title("Absolute error with Euler Method")
plt.show()




plt.subplot(2,1,2)
plt.plot(x_RK,error_RK)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["Absolute Error"])
plt.title("Absolute error with Runge-Kutta Method")
plt.show()
```

$\text{Exercise 2}$

1. write a python code to solve the following linear equations


\begin{equation}
  \begin{cases}
    x+y+z+w  = 13\\
    2x+3y-w = -1\\
    -3x+4y+z+2w  = 10\\
    x+2y-z+w = 1
  \end{cases}.
\end{equation}


```python
def solve(v):
    x = v[0]
    y = v[1]
    w = v[2]
    z = v[3]
    
    
    F =  np.empty(4)
    
    F[0] = x + y + z + w -13
    F[1] = 2*x +3*y -w +1 
    F[2] = -3*x + 4*y +z +2*w -10
    F[3] = x + 2*y -z +w -1
    
    return F



v =  np.array([1,1,1,1])


print(fsolve(solve,v))
```

2. write a python code to solve the following nonlinear equations
\begin{equation}
\begin{cases}
    x^2 + y + x = 2\\
    2e^x + 3y = 8
\end{cases}
\end{equation}

```python
def solver(v):
    x = v[0]
    y = v[1]
    
    
    F =  np.empty(2)
    
    F[0] = x**2 + y +x -2
    F[1] = 2*np.exp(x) + 3*y -8
    
    return F



v =  np.array([1,1])


print(fsolve(solve,v))
```
