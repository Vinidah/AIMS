```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n  = 20
p = 0.1
np.random.binomial(n,p)
```

```python
sns.distplot(np.random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False, label='binomial')
sns.distplot(np.random.poisson(10 *0.5, size=1000), hist=True, kde=False, label='poisson')
sns.distplot(np.random.normal(10 * 0.5,10 * 0.5, size=10), hist=True, kde=False, label='normal')
plt.show() 
```
