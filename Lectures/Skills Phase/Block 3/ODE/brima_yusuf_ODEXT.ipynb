```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t,y):
    y1,y2 = y
    dy1dt = 3*t - 2*y1
    dy2dt = 2*t - 2*y2
    
    return np.array([dy1dt,dy2dt])
```

```python
def ModifiedEuler(f,c,t):
    
    h = t[1] - t[0]
    
    N = len(t)
    M  = len(c)
    
    z = np.zeros((N,M))
    z[0] = c
    
    for n in range(0,N-1):
        k1 =  h *f(t[n],z[n])
        k2 =  h * f(t[n] + h , z[n] + k1)
        z[n+1] =  z[n] +  0.5 * (k1 + k2)
    return z

def exact(t,c):
    return np.array([c[0] * np.exp(t),c[1] * np.exp(t)])
```

```python
t =  np.linspace(0,1,100)
c =  np.array([1,0])
z = ModifiedEuler(f,c,t)
c = np.array([2,1])

exact_result = exact(t,c)
```

```python
len(exact_result[0])
```

```python
fig = plt.figure(1,figsize=(24.4,15.6))
plt.subplot(2,2,1)
plt.plot(t,z[:,0])
plt.plot(t[::10],exact_result[0][::10],"ro")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["Modified Euler","exact"])
plt.title("x1 plot")



plt.subplot(2,2,2)
plt.plot(t,abs(z[:,0]- exact_result[0]))
plt.xlabel("x")
plt.ylabel("Error")
plt.legend(["Approx error"])
plt.title("Approximation Error")

plt.subplot(2,2,3)
plt.plot(t,z[:,1])
plt.plot(t[::10],exact_result[1][::10],"ro")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["Modified Euler","exact"])
plt.title("x1 plot")



plt.subplot(2,2,4)
plt.plot(t,abs(z[:,1]- exact_result[1]))
plt.xlabel("x")
plt.ylabel("Error")
plt.legend(["Approx error"])
plt.title("Approximation Error")
```
